# Dimensiones de los datos


1. ````scalar````: Es un valor único que no está integrado en una lista
2. ````vector````: Es una lista, un array de 1 dimensión
3. ````matriz````: Es una arreglo de datos de 2 dimensiones. Es una tabla con columnas y filas
4. ````Tensor````: Es una arreglo de datos de 3 o más dimensiones.

## Comandos
* ````ndim```` -> Indica la cantidad de dimensiones de un array
* ````ndmin```` -> Con un bandera, ayuda a agregar dimensiones de un array desde su creación.
* ````np.expand_dims```` -> Junto con np, expande la dimensión del array

`````python
scalar = np.array(42)
print(scalar)
scalar.ndim
`````

`````python
vector = np.array([3,4,5])
print(vector)
vector.ndim
`````

`````python
matriz = np.array([[3,4,5],[8,9,7]])
print(matriz)
matriz.ndim
`````

`````python
matriz_2 = np.array([[3,4,5],[8,9,7],[2,5,1],[86,94,75]])
print(matriz_2)
matriz_2.ndim
`````

`````python
tensor = np.array([[[3,4,5],[8,9,7],[2,5,1],[86,94,75]],[[3,4,5],[8,9,7],[2,5,1],[86,94,75]]])
print(tensor)
tensor.ndim
`````

## Agregar o eliminar dimensiones

### Agregar dimensiones desde la creación del array.

`````python
tensor2 = np.array([1,2,3],ndmin=10)
print(tensor2)
tensor2.ndim
`````
